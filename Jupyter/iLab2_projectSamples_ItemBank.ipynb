## Perform aspect term detection and then predict sentiment

````python
import pandas as pd
import numpy as np
````

````python
df_items = pd.read_csv('Data/SFS Item Bank - iLab.csv', encoding='latin')
````

````python
df_items.info()
````

````python
df_items.head()
````

````python
from pyabsa.functional import ABSADatasetList
from pyabsa.functional import ATEPCCheckpointManager

df_items.Text = df_items.ID.astype(str) + ' ' + df_items.Text
inference_source = list(df_items.Text)
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(checkpoint='multilingual-256-2')
atepc_result = aspect_extractor.extract_aspect(
    inference_source=inference_source,
    save_result=True,
    # print_result=True,  # print the result
    pred_sentiment=True,  # Predict the sentiment of extracted aspect terms
)
````

````python
df_out = pd.DataFrame({'Item_ID':[],'aspect':[],'sentiment':[],'confidence':[]})
df_item = pd.DataFrame({'Item_ID':[],'Item_Text':[]})

for ex_id, result in enumerate(atepc_result):
    item_id, comment = result['sentence'].split(' ', 1)
    text_printing = comment  # result['sentence']
    for i in range(len(result['aspect'])):
        aspect_info = '<b>{}&nbsp;</b>'.format(
            result['aspect'][i],
        )
        text_printing = text_printing.replace(result['aspect'][i], aspect_info)

        dict_out = {'Item_ID': [item_id], 'aspect': [result['aspect'][i]],'sentiment':[result['sentiment'][i]],'confidence':[result['confidence'][i]]}
        df_out = pd.DataFrame(dict_out)
        df_out.to_sql(con=database_connection, name='aspect_Item_Bank', if_exists='append', index=False)
    df_item = pd.DataFrame({'Item_ID':[item_id], 'Item_Text':[text_printing[0:4000]]})
    df_item.to_sql(con=database_connection, name='fmt_Item_Bank', if_exists='append', index=False)
````
